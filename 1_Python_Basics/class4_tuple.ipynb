```python
# pack and unpacking tuple
pack_tuple=1,2,"abc",True,1.23
print(pack_tuple)
# unpack_tuple
n1,n2,s,b,f=pack_tuple
print(n1,n2,s,b,f)
```

```python
# unpacking with *
tpl=(1,2,3,4,5,6,7,8)
first,*mid,m=tpl

print(first)
print(mid)
print(m)
```

```python
# nestd tuple

n_tuple=((1,2),("a","b"),(True,False),[1,2,3])
for i in n_tuple:
    for j in i:
        print(j)
print(n_tuple[3])
```
