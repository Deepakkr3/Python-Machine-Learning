```python
# Nested dict
nested_dic={
    "student1":
    {
        "name":"moh",
        "age":43
    }
    ,
    "student2":
    {
        "name":"vishal",
        "age":43
    }
}

print(nested_dic)

print(nested_dic["student1"]["name"])
```

```python
# itrating over nested dict
for st,it in nested_dic.items():
    print(st,it["name"])
```

```python
# dict compherenhension

sqr={ i:i*i for i in range(5)}
print(sqr)
```

```python
sqr={ i:i*i for i in range(5) if i%2 ==0}
print(sqr)
```

```python
numbers=[1,1,2,3,1,5,5,3,8,9,6,5]
friq={}
for it in numbers:
    if it in friq:
        friq[it]=friq[it]+1
    else:
        friq[it]=1
print(friq)
```
