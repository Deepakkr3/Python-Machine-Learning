```python
# map function apply to the all list

def square(x):
    return x*x
square(3)

```

```python
numbers=[1,2,3,4,5,6,7]

x=list(map(square,numbers))
print(x)
```

```python
y=list(map( lambda x:x*x,numbers))
print(y)
```

```python
numbers1=[1,2,3]
number2=[4,5,6]
add_number=list(map(lambda x,y:x+y,numbers1,number2))
print(add_number)
```
