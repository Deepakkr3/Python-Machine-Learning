```python
# variable length args
# positional and keyword args
def prin_numbers(*args):
    for n in args:
        print(n)
prin_numbers(1,2,"ram",True)
```

```python
# keyword args
def print_details(**args):
    for k,v in args.items():
        print(k,v)
print_details(name="deepak",age=10)
```

```python
def pos_key(*args,**keyarg):
    for k in args:
        print("pos arg",k)
    print("key args_________")
    for key,val in keyarg.items():
        print(key,val)


    
pos_key(1,2,3,12,22,name="deepak")
 
```
