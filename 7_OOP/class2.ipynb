```python
# inharitance
class Car:
    def __init__(self,window,type):
        self.window=window
        self.car_type=type
    def drive(self):
        print("person deiving the car type",self.car_type,self.__class__)


car1=Car(4,"petrol");
car1.drive()

        
```

```python
class Tesla(Car):
    def __init__(self, window, type,is_self_deive):
        # super().__init__(window, type)
        Car.__init__(self,window,type)
        self.self_drive=is_self_deive

    def drive(self):
        print("person deiving the car type",self.car_type,self.__class__)
        print("is self drive car",self.self_drive)


    

tesla_car=Tesla(3,"kvk",True)
print(tesla_car)
tesla_car.drive()

```

```python
# multiple inharitance

class Animal:
    def __init__(self,name):
        self.animal_name=name

    def animal_func(self):
        print(f"animal name is {self.animal_name}")


class Pet:
    def __init__(self,petname):
        self.pet_name=petname
    def pet_fun(self):
        print(f"pet name is {self.pet_name}")


class Dog(Animal,Pet):
    def __init__(self, animalname,petname):
        Animal.__init__(    self,animalname)
        Pet.__init__(self,petname)


dog2=Dog("animal name DOG","pet name DUGG")
dog2.animal_func()
dog2.pet_fun()
```
