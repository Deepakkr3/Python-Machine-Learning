```python
class Dog:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def greet(self):
        print(f"{self.name} with age {self.age}")
        


dog1=Dog("A",23)
print(dog1.name)
print(dog1.age)
dog1.greet()
        
```

```python
# class with instance method

class Dog:
    def __init__(self,name,age):
        self.name=name
        self.age=age

    def intro(self):
        print("intro func called",self.name,self.age)

d1=Dog("ABC",23)
d1.intro()

        
```
