```python
# multiple action in same form 

class Parrent:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def greet(self):
        print(f"greet method has been called {self.name} {self.age}")
    
    
class Child(Parrent):
    def __init__(self, name, age,education):
        Parrent.__init__(self,name,age)
        self.education=education

    def greet(self):
        print(f"greet method has been called {self.name} {self.age} education {self.education}")


par=Parrent("dee",24)
par.greet()
ch=Child("dee",43,"b tech")
ch.greet()
        
```

```python
from abc import ABC,abstractclassmethod

class Car(ABC):
    @abstractclassmethod
    def aboutVahcal(self):
        pass
class Maruti(Car):
    def __init__(self,name):
        self.carname=name
    def aboutVahcal(self):
        print(f"about car name is {self.carname}")
m=Maruti("maruti")
m.aboutVahcal()
```
